```python
import torch
from transformers import pipeline
from transformers import GPT2LMHeadModel, GPT2Tokenizer
```

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
device
```

```python
# initializing tokenizer and model from pretrained GPT2 model
tokenizer = GPT2Tokenizer.from_pretrained('gpt2')
model = GPT2LMHeadModel.from_pretrained('gpt2', pad_token_id=tokenizer.eos_token_id)
```

```python
sentence = "What is Machine Learning?"
```

```python
# encoding sentence for model to process
inputs = tokenizer.encode(sentence, return_tensors='pt')
```

```python
# generating text
outputs = model.generate(inputs, max_length=200, do_sample=True, num_beams=5, no_repeat_ngram_size=2, early_stopping=True)
```

```python
# decoding text
text = tokenizer.decode(outputs[0], skip_special_tokens=True)
# printing output
print(text)
```

```python
from transformers import AutoTokenizer, AutoModelForCausalLM
```

```python
model_name = 'gpt2'
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python
model = AutoModelForCausalLM.from_pretrained(model_name).to(device)
```

```python
input_text = "I am really impressed"
max_length = 128

input_ids = tokenizer(input_text, return_tensors='pt')
```

```python
input_ids
```

```python
input_ids = input_ids['input_ids'].to(device)
```

```python
output = model.generate(input_ids, max_length=max_length, num_beams=5, do_sample=False)
```

```python
output
```

```python
tokenizer.decode(output[0])
```

```python
output = model.generate(input_ids, max_length=max_length, num_beams=5, do_sample=False, no_repeat_ngram_size=2)
```

```python
tokenizer.decode(output[0])
```
